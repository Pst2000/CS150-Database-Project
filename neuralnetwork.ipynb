```python
import pandas as pd
import numpy as np
```

```python
# Get the data:
train_filepath = 'data/train.csv'
traindata = pd.read_table(train_filepath)
```

```python
from sklearn.utils import shuffle
traindata = shuffle(traindata)
```

```python
traindata.head(10)
```

```python
SID = np.unique(traindata['Anon Student Id'])
```

```python
Step = np.unique(traindata['Step Name'])
```

```python
KC = []
for _,row in traindata.iterrows():
    if pd.isnull(row['KC(Default)']):
        continue
    KC.extend(row['KC(Default)'].split("~~"))
```

```python
Problem = np.unique(traindata['Problem Name'])
```

```python
len(Problem)
```

```python
KC = np.unique(KC)
```

```python
SID_dic = {}
Step_dic = {}
KC_dic = {}
Problem_dic = {}
for sid in SID:
    SID_dic[sid] = [0, 0]
for step in Step:
    Step_dic[step] = [0, 0]
for kc in KC:
    KC_dic[kc] = [0, 0]
for problem in Problem:
    Problem_dic[problem] = [0, 0]
```

```python
# calculate the tuple(probabilities) of each CFA
for _,row in traindata.iterrows():
    sid = row['Anon Student Id']
    step = row['Step Name']
    problem = row['Problem Name']
    kc_list = []
    if not pd.isnull(row['KC(Default)']):
        kc_list = row['KC(Default)'].split("~~")
    
    SID_dic[sid][1] += 1
    Step_dic[step][1] += 1
    Problem_dic[problem][1] += 1
    if kc_list:
        for kc in kc_list:
            KC_dic[kc][1] += 1
            
    if row['Correct First Attempt'] == 1:
        SID_dic[sid][0] += 1
        Step_dic[step][0] += 1
        Problem_dic[problem][0] += 1
        if kc_list:
            for kc in kc_list:
                KC_dic[kc][0] += 1
```

## Laplace Smoothing

```python
for key in Step_dic.keys():
    Step_dic[key][0] += 2
    Step_dic[key][1] += 2
```

```python
# specific for training set!!
Final_list = []
for _,row in traindata.iterrows():
    sid = row['Anon Student Id']
    step = row['Step Name']
    problem = row['Problem Name']
    kc_list = []
    if not pd.isnull(row['KC(Default)']):
        kc_list = row['KC(Default)'].split("~~")
        
    P_sid = SID_dic[sid][0]/SID_dic[sid][1]
    P_step = Step_dic[step][0]/Step_dic[step][1]
    P_problem = Problem_dic[problem][0]/Problem_dic[problem][1]
    P_kc = 1
    if kc_list:
        for kc in kc_list:
            P_kc *= KC_dic[kc][0]/KC_dic[kc][1]
    else:
        # KC is NaN
        P_kc = 0.8
    Final_list.append([P_sid, P_step, P_kc, P_problem])
```

# The following are for the network proportion

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np

train_feat = np.load("train_feat_2.npy")
val_feat = np.load("val_feat_2.npy")
train_label = np.load("train_label_2.npy")
val_label = np.load("val_label_2.npy")

new_train_label = []
for i in train_label:
    new_train_label.append([i])

new_val_label = []
for i in val_label:
    new_val_label.append([i])
    
X = torch.Tensor(train_feat)
y = torch.Tensor(new_train_label)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(4,8)
        self.fc2 = nn.Linear(8,16)
        self.fc3 = nn.Linear(16,4)
        self.fc4 = nn.Linear(4,1)
        
    def forward(self,x):
        x = self.fc1(x)
        x = F.tanh(x)
        x = self.fc2(x)
        x = F.tanh(x)
        x = self.fc3(x)
        x = F.tanh(x)
        x = self.fc4(x)
        return x

model = Net()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
for param in model.parameters():
    print(param)
# model.parameters()
```

```python
model.load_state_dict(torch.load("model.pth"))
model.eval()
```

```python
from tqdm import tqdm

epochs = 1000
for i in tqdm(range(epochs)):
    y_pred = model(X)
#     print(y_pred)
#     break
    loss = criterion(y_pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

# Test on given test set

```python
test_list = []
test_label = []
for _,row in testdata.iterrows():
    # choose the test set
    if not pd.isnull(row['Correct First Attempt']):
        sid = row['Anon Student Id']
        step = row['Step Name']
        problem = row['Problem Name']
        kc_list = []
        if not pd.isnull(row['KC(Default)']):
            kc_list = row['KC(Default)'].split("~~")

        P_sid = SID_dic[sid][0]/SID_dic[sid][1]
        if problem not in Problem_dic.keys():
            P_problem = Problem_avg
        else:
            P_problem = Problem_dic[problem][0]/Problem_dic[problem][1]
            
        if step not in Step_dic.keys():
            P_step = Step_avg
        else:
            P_step = Step_dic[step][0]/Step_dic[step][1]
        P_kc = 1
        if kc_list:
            for kc in kc_list:
                P_kc *= KC_dic[kc][0]/KC_dic[kc][1]
        test_list.append([P_sid, P_step, P_kc, P_problem])
        test_label.append(row['Correct First Attempt'])
```

```python
final_CFA_list = []
for _,row in testdata.iterrows():
    if pd.isnull(row['Correct First Attempt']):
        sid = row['Anon Student Id']
        step = row['Step Name']
        problem = row['Problem Name']
        kc_list = []
        if not pd.isnull(row['KC(Default)']):
            kc_list = row['KC(Default)'].split("~~")

        P_sid = SID_dic[sid][0]/SID_dic[sid][1]
        if problem not in Problem_dic.keys():
            P_problem = Problem_avg
        else:
            P_problem = Problem_dic[problem][0]/Problem_dic[problem][1]
            
        if step not in Step_dic.keys():
            P_step = Step_avg
        else:
            P_step = Step_dic[step][0]/Step_dic[step][1]
        P_kc = 1
        if kc_list:
            for kc in kc_list:
                P_kc *= KC_dic[kc][0]/KC_dic[kc][1]
        
        input_X = [P_sid, P_step, P_kc, P_problem]
        X = torch.Tensor(input_X)
        y = model.forward(X)
        if y > 0.54:
            final_CFA_list.append(1)
        else:
            final_CFA_list.append(0)
    else:
        final_CFA_list.append(row['Correct First Attempt'])
```
